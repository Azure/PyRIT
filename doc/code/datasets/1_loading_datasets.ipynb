# 1. Loading Built-in Datasets

PyRIT includes many datasets to help you get started. While it is not a goal to define harms or objectives within PyRIT, it **is** a goal to make it as easy as possible to include your own or those defined by others. And many are included out of the box.

One note is that datasets are better managed as part of [memory](../memory/8_seed_database.ipynb). In memory, the data is normalized and easy to query. But this is a way to get started, and these are easily added to the database. 

The following command shows how you can list all the builtin Datasets available for this purpose. Some of these are local and some are fetched remotely.

```python
from pyrit.datasets import SeedDatasetProvider

SeedDatasetProvider.get_all_dataset_names()
```

You can retrieve all built-in datasets by using `SeedDatasetProvider.fetch_all_datasets()`, or you can retrieve specific ones using the dataset name (which will return a list of datasets).

```python
datasets = await SeedDatasetProvider.fetch_all_datasets(dataset_names=["airt_illegal", "airt_malware"])

for dataset in datasets:
    for seed in dataset.seeds:
        print(seed.value)
```

Although covered in more detail in the [memory](../memory/0_memory.md) section, these prompts can easily be added to PyRIT memory, where it is easier to manage them (e.g. query by harm category, etc.).

```python
from pyrit.memory import CentralMemory
from pyrit.setup.initialization import IN_MEMORY, initialize_pyrit

initialize_pyrit(memory_db_type=IN_MEMORY)

memory = CentralMemory().get_memory_instance() 
await memory.add_seed_datasets_to_memory_async(datasets=datasets, added_by="pyrit")

# Memory has flexible querying capabilities
memory.get_seeds(harm_categories=["illegal"], is_objective=True)
```
